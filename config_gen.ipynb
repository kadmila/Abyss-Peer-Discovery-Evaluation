```python
import json
```

```python
city_abbr = {}
city_radius = {}
city_population = {}
network_stats = {}

with open('./city_abbr.json') as f_city_abbr:
    city_abbr = json.load(f_city_abbr)

with open('./city_radius.json') as f_city_radius:
    city_radius = json.load(f_city_radius)

with open('./city_population.json') as f_city_population:
    city_population = json.load(f_city_population)

with open('./network_stats.json') as f_network_stats:
    network_stats = json.load(f_network_stats)
```

```python
combined_data = {}

i = 12
for city, abbr in city_abbr.items():
    combined_data[abbr] = {
        "city": city,
        "radius": city_radius[city],
        "population": city_population[city],
        "network_stats": {},
        "ip_prefix": i
    }
    i += 4

for city_to_city, stat in network_stats.items():
    city1, city2 = city_to_city.split('|')
    city1_abbr = city_abbr[city1]
    city2_abbr = city_abbr[city2]
    combined_data[city1_abbr]["network_stats"][city2_abbr] = stat
    combined_data[city2_abbr]["network_stats"][city1_abbr] = stat
```

```python
def list_sub(a: list, b: list) -> list:
    return [v for v in a if a not in b]

result_cities = []
is_set = False
for k, v in combined_data.items():
    g_list = [k]
    g_list.extend(v['network_stats'].keys())
    
    if not is_set:
        result_cities = g_list
        is_set = True
    
    rm = list_sub(result_cities, g_list)
    result_cities = list_sub(result_cities, rm)

result = {}
for res_c in result_cities:
    result[res_c] = combined_data[res_c]
```

```python
with open('./mininet-control/city_config.json', 'w') as f_city_config:
    json.dump(result, f_city_config, indent=4)
```
